### Reproducing Figure 4c

1. go to the `scripts/simulated-annealing-monte-carlo/` directory and run:

   ```bash
   sa_desktop.py
   ```

2. Running this script generates four Excel output files. One of these files is:

   ```
   200n_e4_acceptance_rates.xlsx
   ```

3. Open `scripts/simulated-annealing-monte-carlo/acceptance_rate.py` and update the `file1_path` variable so that it points to the location of your generated Excel file. For example:

   ```python
   file1_path = "/path/to/your/200n_e4_acceptance_rates.xlsx"
   ```

   Replace the example path with the actual location of the file on your system.

4. Run `acceptance_rate.py`.

   The script reads the acceptance-rate data from the Excel file and generates the plot shown in **Figure 4c**.
