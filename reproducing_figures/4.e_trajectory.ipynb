### Reproducing Figure 4e

Run the following script:

```text
scripts/simulated-annealing-monte-carlo/plot_trajectory.py
```

This script runs a single simulated annealing trajectory and generates the trajectory plot shown in **Figure 4e**.

#### Required Input

The script requires a metadata (`.tsv`) file containing the barcode information. You can use the same file used in our experiments:

```text
data/generated_barcodes.tsv
```

Update the `tsv_file` parameter in `plot_trajectory.py` so that it points to the location of this file on your system.

```python
if __name__ == "__main__":
    accepted_states, trajectory_coords, trajectory_energies = run_single_spatial_simulated_annealing(
        tsv_file="/path/to/generated_barcodes.tsv",
        k_neighbors=150,
        spatial_weight=0.6,
        energy_weight=0.4,
        outer_iterations=200,
        inner_iterations=50,
        skip_points_for_plot=5
    )
```

#### Adjustable Parameters

You can modify several parameters to explore different behaviors of the algorithm.

**Spatial and energy weights**

These parameters control the relative contribution of the spatial distance and energy terms in the objective function.

```python
spatial_weight=0.6,
energy_weight=0.4,
```

**Number of nearest neighbors**

This parameter determines how many neighboring barcodes are considered during the optimization.

```python
k_neighbors=150,
```

**Trajectory smoothing**

To improve the readability of the trajectory plot, the script connects every five accepted states when drawing the trajectory.

```python
skip_points_for_plot=5
```

Increasing this value produces a smoother trajectory, while decreasing it shows more of the individual optimization steps.

> **Important:** Changing any of the parameters above modifies the simulated annealing algorithm and therefore changes the resulting trajectory and the generated figure. The published Figure 4e was generated using the parameter values shown above.
